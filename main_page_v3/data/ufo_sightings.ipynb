```python
import pandas as pd
import numpy as np
import io
from scipy import stats
from google.colab import files
uploaded = files.upload()
```

```python
ufo_sightings = pd.read_csv(io.BytesIO(uploaded['kaggle_ufo_sightings.csv']))
ufo_sightings_df = ufo_sightings.copy()
```

```python
ufo_sightings_df.head()
```

```python
ufo_sightings_df = ufo_sightings_df.dropna()
#ufo_sightings_df
```

```python
ufo_sightings_df['country'].value_counts()
```

```python
ufo_sightings_df = ufo_sightings_df.loc[ufo_sightings_df['country']=='us']
#ufo_sightings_df
```

```python
ufo_sightings_df['date'] = ufo_sightings_df['datetime'].str[:-6]
ufo_sightings_df['date'] = ufo_sightings_df['date'].str[-4:]
ufo_sightings_df['time'] = ufo_sightings_df['datetime'].str[-5:]
```

```python
#ufo_sightings_df
```

```python
ufo_sightings_df = ufo_sightings_df.sort_values('date', axis=0)
ufo_sightings_df
```

```python
ufo_sightings_df = ufo_sightings_df.loc[ufo_sightings_df['date'].str[0]=='1']
#ufo_sightings_df
```

```python
shapes = ufo_sightings_df['shape'].unique()
print(shapes)
```

```python
ufo_sightings_df['shape'].value_counts()
```

```python
shapes_incl = ['triangle', 'circle', 'disk']
shapess = [shape for shape in shapes if shape not in shapes_incl]
print(shapess)
```

```python
temp_df = ufo_sightings_df

for shape in shapess:
  temp_df = temp_df.loc[temp_df['shape']!= shape]
  
```

```python
temp_df
```

```python
to_drop = ['city', 'duration (hours/min)', 'date posted', 'country']
temp_df = temp_df.drop(columns = to_drop)
temp_df
```

```python
temp_df = temp_df.loc[temp_df['date'].str[2]=='9']
temp_df
```

```python
temp_df = temp_df.loc[temp_df['state']!='ak']
```

```python
temp_df = temp_df.loc[temp_df['state']!='hi']
```

```python
temp_df.to_csv('ufo.csv')
files.download('ufo.csv')
```
